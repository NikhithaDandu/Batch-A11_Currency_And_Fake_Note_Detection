```python
import numpy
import tkinter as tk
from tkinter import filedialog
from tkinter import *
from PIL import ImageTk, Image
import tensorflow as tf
from tensorflow.keras.models import load_model
from detecttry import impli
from class_new import *

root = tk.Tk()
root.title('Browser')
root.resizable(0,0)

file_path = r"C:/Users/divya/Desktop/fake_note_project/bg_img.jpg"
im = Image.open(file_path)
im = im.resize((800,580))
bg_image = ImageTk.PhotoImage(im)
w = bg_image.width()
h = bg_image.height()
root.geometry("%dx%d+50+30" % (w, h))
cv = tk.Canvas(width=w, height=h)
cv.pack(side='top', fill='both', expand='yes')
cv.create_image(0, 0, image=bg_image, anchor='nw')
cv.create_text(w//4, 110, text="CURRENCY AND FAKE NOTE DETECTION", fill="maroon", anchor='nw', font=('lucida',15,'bold'))

sign_image = Label(root)
                        
def res():
    global bg,result,top
    file = r"C:/Users/divya/Desktop/fake_note_project/bg_img.jpg"
    im = Image.open(file)
    im = im.resize((300,150))
    bg = ImageTk.PhotoImage(im)
    result = impli(file_path)    
    top=Toplevel()
    top.title("result")
    top.geometry('300x150')
    top.resizable(0,0)
    my_lab = Label(top,image=bg)
    my_lab.place(x=0,y=0,relwidth=1,relheight=1)
    label = Label(top,text=result,bg='white',font=('lucida',10,'bold'))
    label.place(relx=0.5,rely=0.5,anchor='center')
    
    
def val():
    global bg1,top1
    file1 = r"C:/Users/divya/Desktop/fake_note_project/bg_img.jpg"
    im1 = Image.open(file1)
    im1 = im.resize((300,150))
    bg1 = ImageTk.PhotoImage(im1)
    res = impli(file_path)
    if(res=='Original Note'):
        value = pred(file_path)
    else:
        value="Fake Note"
    top1 = Toplevel()
    top1.title("value")
    top1.geometry('300x150')
    top1.resizable(0,0)
    my_lab = Label(top1,image=bg1)
    my_lab.place(x=0,y=0,relwidth=1,relheight=1)
    lab = Label(top1,text=value,bg='white',font=('lucida',10,'bold'))
    lab.place(relx=0.5,rely=0.5,anchor='center')
    
    
def upload_image():
    try:
        global file_path
        file_path=filedialog.askopenfilename()
        uploaded=Image.open(file_path)
        uploaded = uploaded.resize((350,200))
        s_im=ImageTk.PhotoImage(uploaded)
        sign_image.configure(image=s_im)
        sign_image.image=s_im
        cv.create_image(220, 190, image=s_im, anchor='nw')
        top.destroy()
        top1.destroy()
    except Exception as e:
        print(e)
        pass
    
upload = tk.Button(cv, text="Upload Image",bg='gray',fg='black',command=upload_image)
upload.pack(side=TOP,pady=150)
r = tk.Button(cv, text="fake/original",bg='gray',fg='black',command=res)
r.pack(side=LEFT,padx=180,pady=30)
v = tk.Button(cv, text="value",bg='gray',fg='black',command=val)
v.pack(side=RIGHT,padx=150)
root.mainloop()
```
